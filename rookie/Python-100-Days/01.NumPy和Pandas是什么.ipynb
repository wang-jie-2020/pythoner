### 一、numpy 和 pandas 是什么</br>

两个都是计算函数包(数学公式 和 常见算法), 前者抽象一些也更基础一些, 核心对象是ndarray. pandas 对二维数组humanize一些, 核心对象是dataframe(series). </br>

**numpy: ndarray = array, pandas: series = ndarray + index, dataframe = n * series**</br>

```python
import numpy as np
import pandas as pd

array = np.random.randint(60, 101, (5, 3))  # 5行3列的二维数组

# ndarray
ndarray = np.array(array)

# index: (行)索引
# data: 限制是一维数组
# series: 一时想不到怎么描述它, 键值对?
series = pd.Series(array[:, 0], index=['狄仁杰', '白起', '李元芳', '苏妲己', '孙尚香'])

# dataframe 数据表格
dataframe = pd.DataFrame(data=array, columns=['语文', '数学', '英语'], index=['狄仁杰', '白起', '李元芳', '苏妲己', '孙尚香'])

ndarray, series, dataframe
```

### 三、DataFrame

- `Series`：表示一维的数据。跟一维数组的区别在于每条数据都有对应的索引，处理数据的方法比`ndarray`更为丰富。
- `DataFrame`：表示二维的数据。跟二维数组相比，`DataFrame`有行索引和列索引，而且提供了100+方法来处理数据。
- `Index`：为`Series`和`DataFrame`提供索引服务。


```python
import numpy as np
import pandas as pd

ser1 = pd.Series(data=[120, 380, 250, 360], index=['一季度', '二季度', '三季度', '四季度'])
ser2 = pd.Series({'一季度': 320, '二季度': 180, '三季度': 300, '四季度': 405})
ser1, ser2
```

```python
stu_names = ['狄仁杰', '白起', '李元芳', '苏妲己', '孙尚香']
cou_names = ['语文', '数学', '英语']
scores_arr = np.random.randint(60, 101, (5, 3))
df1 = pd.DataFrame(data=scores_arr, columns=cou_names, index=stu_names)
df1
```
